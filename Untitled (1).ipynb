```python
# ▓▓ 0 ▓▓  install once per Colab runtime
!pip install -qU "pennylane-lightning[gpu]" scikit-learn matplotlib seaborn

# ▓▓ 1 ▓▓  imports & helper
import time, seaborn as sns, matplotlib.pyplot as plt, pennylane as qml
from pennylane import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix)

def device_qgpu(n_data, spare=1):
    try:  return qml.device("lightning.gpu", wires=n_data+spare)
    except qml.DeviceError:
          return qml.device("default.qubit", wires=n_data+spare)

# ▓▓ 2 ▓▓  balanced splits  (train 600, val 120, rest test)
X, y_all = fetch_openml("Fashion-MNIST", version=1, return_X_y=True, as_frame=False)
y_all    = y_all.astype(int)
pos, neg = np.where(y_all==7)[0], np.where(y_all!=7)[0]

rng = np.random.default_rng(0)
train_idx = np.concatenate([rng.choice(pos,300,replace=False),
                            rng.choice(neg,300,replace=False)])
mask = np.ones(len(X),bool); mask[train_idx]=False
X_train, y_train = X[train_idx], np.concatenate([np.ones(300), np.zeros(300)])

# validation 60 + 60
pos_left = np.where(mask & (y_all==7))[0]
neg_left = np.where(mask & (y_all!=7))[0]
val_idx  = np.concatenate([rng.choice(pos_left,60,replace=False),
                           rng.choice(neg_left,60,replace=False)])
X_val, y_val = X[val_idx], np.concatenate([np.ones(60), np.zeros(60)])
mask[val_idx] = False

# remaining → test
X_test, y_test = X[mask], (y_all==7)[mask].astype(float)

print(f"sizes  train={len(X_train)}  val={len(X_val)}  test={len(X_test)}")

# PCA→4  &  scale [0,π]
pca   = PCA(4).fit(X_train)
scaler= MinMaxScaler((0,np.pi)).fit(pca.transform(X_train))
X_train = scaler.transform(pca.transform(X_train)); y_train = np.array(y_train)
X_val   = scaler.transform(pca.transform(X_val))
X_test  = scaler.transform(pca.transform(X_test))

# ▓▓ 3 ▓▓  circuit (4 qubits + 1 spare)
n_q  = 4
dev  = device_qgpu(n_q)

@qml.qnode(dev, interface="autograd")
def circuit(w,x):
    for i in range(n_q):
        qml.RY(x[i], wires=i)
    qml.templates.StronglyEntanglingLayers(w, wires=range(n_q))
    return qml.expval(qml.PauliZ(0))

metric_diag = qml.metric_tensor(circuit, approx="diag")   # P×P

# ▓▓ 4 ▓▓  BCE loss (POS_W = 1  → neutral)
def bce(w):
    p=np.stack([(circuit(w,x)+1)/2 for x in X_train])
    eps=1e-8
    return -np.mean(y_train*np.log(p+eps)+(1-y_train)*np.log(1-p+eps))
grad_fn=qml.grad(bce)

# ▓▓ 5 ▓▓  QNG training  (diag, mini-batch)
LAYERS=3
shape =(LAYERS,n_q,3)
P     =int(np.prod(shape))
w     =np.random.normal(0,0.1,size=shape,requires_grad=True)

EPOCHS, ETA, LAM, BATCH = 20, 0.2, 1e-2, 60
loss=[]
for ep in range(1,EPOCHS+1):
    g=grad_fn(w).ravel()
    idx=rng.choice(len(X_train),BATCH,replace=False)
    Gv=np.zeros(P)
    for i in idx:
        Gv += np.diag(metric_diag(w,X_train[i]).reshape(P,P))
    Gv = Gv/len(idx)+LAM
    w  -= ETA*(g/Gv).reshape(shape)
    loss.append(float(bce(w)))
    print(f"epoch {ep:2d}  loss {loss[-1]:.4f}")

# ▓▓ 6 ▓▓  choose threshold τ via F_{β=0.5}
β=0.5
p_val=np.array([(circuit(w,x)+1)/2 for x in X_val])
best_tau,best_score=0.5,0
for t in np.linspace(0.05,0.95,19):
    y_tmp=(p_val>t).astype(int)
    prec,rec = precision_score(y_val,y_tmp), recall_score(y_val,y_tmp)
    f_β = (1+β**2)*prec*rec/(β**2*prec+rec+1e-8)
    if f_β>best_score:
        best_score,best_tau=f_β,t
print(f"chosen τ={best_tau:.2f}  (val F0.5={best_score:.3f})")

# ▓▓ 7 ▓▓  evaluate on test
p_test=np.array([(circuit(w,x)+1)/2 for x in X_test])
y_pred=(p_test>best_tau).astype(int)

acc  = accuracy_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)
rec  = recall_score(y_test,y_pred)
f1   = f1_score(y_test,y_pred)
cm   = confusion_matrix(y_test,y_pred)

# plot curve
plt.plot(range(1,EPOCHS+1),loss,marker='o'); plt.grid()
plt.title(f"QNG convergence\nTest  acc={acc*100:.2f}%  prec={prec:.3f}  rec={rec:.3f}  F1={f1:.3f}")
plt.xlabel("epoch"); plt.ylabel("weighted BCE"); plt.show()

sns.heatmap(cm,annot=True,fmt="d",cmap="Blues",
            xticklabels=["non","sneaker"],yticklabels=["non","sneaker"])
plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("Confusion"); plt.show()
```
